```python
import pickle
import pandas as pd
loaded = pickle.load(open('delta_a100_gpu_xid_dataset.pkl', 'rb'))
```

```python
the_day = "2024-03-10"     #its the day asked in question
day_logs = loaded[the_day] #access the day in loaded
xid_list=[]                #reserve an empty list

for logs in day_logs:
    xid_list.append(logs["XID"]) #use a for loop to loop through all value of XID,put them into list
numb_errors = len(xid_list)
print("The number of errors is", numb_errors)
print(xid_list)
dif_xid = set(xid_list)    #save all the different numbers in xid_list
n = len(dif_xid)
print("There are", n, "differnt numbers of xid on March 10th, 2024. They are", dif_xid)
```

```python
rows = [] #reserve an empty list
for i, j in loaded.items():
    for k in j: #loop through list(or the file) 
        row = k.copy() #make a copy of load
        row['Date'] = i #save different datas into this copy
        rows.append(row) #then save it into the final list
df = pd.DataFrame(rows) #load data into a DataFrame object
print(df)
```
